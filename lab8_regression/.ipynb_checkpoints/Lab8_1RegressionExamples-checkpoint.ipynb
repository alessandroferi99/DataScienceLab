```python
from math import sin, pi
import numpy as np
def f1(x):
    return x*sin(x)+2*x
def f2(x):
    return 10*sin(x)+x**2
def f3(x):
    return np.sign(x)*(x**2+300)+20*sin(x)
```

```python
from sklearn.model_selection import train_test_split
tr = 20
n_samples = 100
X = np.linspace(-tr, tr, n_samples)
y = np.array(list(map(f1,X)))
y1 = np.array(list(map(f1,X)))
y2 = np.array(list(map(f2,X)))
y3 = np.array(list(map(f3,X)))

X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.7, random_state=42, shuffle=True
)
y_test = y_test[X_test.argsort()]
X_test.sort()
```

## Inspection of the functions shape

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(X,y1, label='f1')
plt.plot(X,y2, label='f2')
plt.plot(X,y3, label='f3')
plt.legend()
plt.show()
```

### R2 (R squared)
$R^2=1-\frac{MSE}{std^2}$ close to 1: good predictions, negative or close to 0: wrong predictions
### MAE
$MAE= \frac{1}{n}\sum|y_i-\hat{y_i}|$
### MSE
$MSE= \frac{1}{n}\sum(y_i-\hat{y_i})^2$


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score 
reg = LinearRegression(fit_intercept=True)
X_train = X_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

reg.fit(X_train,y_train)
y_test_pred = reg.predict(X_test)
r2 = r2_score(y_test, y_test_pred)
print(f"R2 score f1: {r2}")
plt.plot(X_test,y_test, label='true')
plt.plot(X_test,y_test_pred, label='predicted')
plt.legend()
plt.show()
X_train, X_test, y_train, y_test = train_test_split(
    X, y2, train_size=0.7, random_state=42, shuffle=True
)
y_test = y_test[X_test.argsort()]
X_test.sort()
reg = LinearRegression(fit_intercept=True)
X_train = X_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

reg.fit(X_train,y_train)
y_test_pred = reg.predict(X_test)
r2 = r2_score(y_test, y_test_pred)
print(f"R2 score f2: {r2}")
plt.plot(X_test,y_test, label='true')
plt.plot(X_test,y_test_pred, label='predicted')
plt.legend()
plt.show()

X_train, X_test, y_train, y_test = train_test_split(
    X, y3, train_size=0.7, random_state=42, shuffle=True
)
y_test = y_test[X_test.argsort()]
X_test.sort()
reg = LinearRegression(fit_intercept=True)
X_train = X_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

reg.fit(X_train,y_train)
y_test_pred = reg.predict(X_test)
r2 = r2_score(y_test, y_test_pred)
print(f"R2 score f3: {r2}")
plt.plot(X_test,y_test, label='true')
plt.plot(X_test,y_test_pred, label='predicted')
plt.legend()
plt.show()
```

The R2 score for f2 is negative, so we have a bad prediction for $f_2(x)$ as we expected:
We have to use Polynomial Regression

## Ridge with GridSearch:

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(
    X, y1, train_size=0.7, random_state=42, shuffle=True
)
y_test = y_test[X_test.argsort()]
X_test.sort()
param_grid = {'alpha': [0.1, 0.2],
              'fit_intercept': [True, False]}

X_train = X_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

reg = Ridge()
gridsearch = GridSearchCV(reg, param_grid, scoring='r2', cv=2)
gridsearch.fit(X_train, y_train)

print(f"best alpha: {gridsearch.best_params_['alpha']}")
print(f"best with fit_intercept = {gridsearch.best_params_['fit_intercept']}")
best_model = gridsearch.best_estimator_
y_test_pred = best_model.predict(X_test)
r2 = r2_score(y_test, y_test_pred)
print(f"R2 score f1: {r2}")

plt.plot(X_test,y_test, label='true')
plt.plot(X_test,y_test_pred, label='predicted')
plt.legend()
plt.show()
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

reg = make_pipeline(PolynomialFeatures(2), LinearRegression())


X_train, X_test, y_train, y_test = train_test_split(
    X, y2, train_size=0.7, random_state=42, shuffle=True
)
y_test = y_test[X_test.argsort()]
X_test.sort()

X_train = X_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

reg.fit(X_train, y_train)
y_test_pred = reg.predict(X_test)
r2 = r2_score(y_test, y_test_pred)
print(f"R2 score f2: {r2}")

plt.plot(X_test,y_test, label='true')
plt.plot(X_test,y_test_pred, label='predicted')
plt.legend()
plt.show()
```
